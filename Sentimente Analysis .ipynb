```python
# Import necessary libraries
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
dataset_path = "C:\\Users\\ARYAN PARIKH\\Downloads\\archive (2)\\Twitter_Data.csv"
df = pd.read_csv(dataset_path)

# Display basic information about the dataset
print("Original Dataset Info:")
print(df.info())

# Sentiment Analysis using TextBlob
df['Sentiment'] = df['clean_text'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)

# Convert sentiment polarity to sentiment labels (positive, negative, neutral)
df['Sentiment_Label'] = df['Sentiment'].apply(lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral'))

# Data Visualization: Plotting a pie chart for sentiment distribution
plt.figure(figsize=(8, 8))
sns.countplot(x='Sentiment_Label', data=df, palette='viridis')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment Label')
plt.ylabel('Count')
plt.show()

# Save the dataset with sentiment analysis results (if needed)
# df.to_csv("twitter_data_with_sentiment.csv", index=False)
```
